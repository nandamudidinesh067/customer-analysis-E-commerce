## Spliting the dataset for Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['Avg. Session Length',
      'Time on App','Time on Website',
      'Length of Membership', ]]
```

```python
Y = df['Yearly Amount Spent']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,
                                                    test_size= 0.3, 
                                                    random_state=10)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
```

```python
ln.fit(X_train, Y_train)
```

```python
coeff = ln.coef_
```

```python
coeffdf = pd.DataFrame(coeff, X.columns, 
                       columns=["Co-Efficients"])
coeffdf
```

## Predictions

```python
prediction = ln.predict(X_test)
```

```python
prediction
```

## Graphical Evaluation

```python
sns.scatterplot(x=prediction, y=Y_test)
plt.xlabel("Predicted Values")
plt.title("Analyzing the Linear Model")
```

## Error Analysis

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
Absolute_error = mean_absolute_error(Y_test, prediction)
Absolute_error
```

```python
Squared_error = mean_squared_error(Y_test, prediction)
Root_Squared = math.sqrt(Squared_error)
Root_Squared
```

## Residual Analysis

```python
residuals = Y_test - prediction
```

```python
sns.displot(residuals, bins=45, kde=True)
```

## Conclusions 

Interpreting the analysis suggests that the most influential factor for clients is their length of membership, rather than the time spent on the app or website. Notably, among the predictors (desktop vs. app), the app has a significantly stronger impact. Surprisingly, the time spent on the desktop website shows no correlation with the amount of money spent

## Interpretation

Two plausible interpretations arise from this data. Firstly, it may indicate that the desktop website requires enhancement to boost customer spending. Alternatively, it suggests that consumers are more swayed by mobile applications, emphasizing the potential for directing efforts towards capitalizing on this trend. Expertise in online marketing is crucial for a nuanced interpretation. The analysis and model effectively prioritize predictor importance
